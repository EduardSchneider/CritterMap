```python
import rdflib
import folium
```

```python
# Finding all species in ontology
g = rdflib.Graph()
g.parse('animaldb.ttl', format='n3')

initquery = g.query("""

PREFIX a: <http://example.com/animal/>
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
select ?speciesname where { 
    ?species rdfs:subClassOf a:Species .
    ?species rdfs:label ?speciesname
}


""")

for row in initquery:
    print(row['speciesname'])
```

```python
import folium
m=folium.Map(location=[51.5,-0.1], zoom_start=5)

# Query all countries, geodata and literal name
countryquery = g.query("""

PREFIX a: <http://example.com/animal/>
PREFIX spif: <http://spinrdf.org/spif#>
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
PREFIX dbo: <http://dbpedia.org/ontology/>
PREFIX geo: <http://www.w3.org/2003/01/geo/wgs84_pos#>
PREFIX owl: <http://www.w3.org/2002/07/owl#>
PREFIX : <http://www.semanticweb.org/eduard/ontologies/2020/9/animal#>

select ?countryname ?lat ?long ?name where { 
    ?country a a:Country .
    ?country rdfs:label ?countryname .
    ?country rdfs:label ?name .
    ?country geo:lat ?lat .
    ?country geo:long ?long .
}
""")

# Query all animals per country, in this case with species being Goose
animalquery = g.query("""

PREFIX a: <http://example.com/animal/>
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
select ?countryname ?animalname where { 
    ?country a a:Country .
    ?country rdfs:label ?countryname .
    ?animal a:isSpecies a:Goose .
    ?animal a:livesIn ?country .
    ?animal rdfs:label ?animalname
} 
""")

# Create bubble markers for each country that has animals of the selected species
for row1 in countryquery:
    tooltip = str(row1['countryname']) + ' has the following animals: '
    for row2 in animalquery:
        if row1['countryname'] == row2['countryname']:
            tooltip += str(row2['animalname']) + ', '
    if tooltip[-2] != ':':
        folium.Circle(
            location = [row1['lat'],row1['long']],
            popup = tooltip,
            radius = 50000,
            color='green',
            fill=True,
            fill_color='green'
        ).add_to(m)
```
